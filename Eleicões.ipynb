```python
# Importando bibliotecas
# (importing libraries)


# bibliotecas para modelagem e matrizes
# (libraries for modeling and matrices)
import basedosdados as bd
import pandas as pd

# Para carregar o dado direto no pandas
# (To load the data directly into pandas)

base_dados = bd.read_table(dataset_id='br_poder360_pesquisas',
                           table_id='microdados',
                           billing_project_id="<YOUR_PROJECT_ID>")
```

```python
# Imprimindo os 10 primeiros registro da base de dados 
# (Printing the first 10 records from the database)

base_dados.head(10)
```

```python
# Criando filtros para a plotagem do gráfico 
# (Creating filters for graph plotting)

filtro =  base_dados[
                    (base_dados.instituto == 'Datafolha') & 
                    (base_dados.ano == 2022) & 
                    (base_dados.sigla_uf.isnull()) & 
                    (base_dados.cargo == 'presidente') &
                    (base_dados.tipo == 'estimulada') &
                    (base_dados.turno == 1) &            
                    ((base_dados.nome_candidato == 'Lula') | (base_dados.nome_candidato == 'Bolsonaro') | 
                    (base_dados.nome_candidato == 'Ciro') | (base_dados.nome_candidato == 'Simone Tebet')) &
                    (base_dados.descricao_cenario.str.contains('cenário 1')==True)
            
            
   ]
```

```python
# Ordenando os dados através da data
# (Sorting data by date)

pesquisas = filtro.sort_values(by='data')
pesquisas
```

```python
# Configurando o backend para plotagem do gráfico
# (Configuring the backend for chart plotting)

pd.options.plotting.backend = "plotly"
```

```python
# Plotagendo o gráfico de linha com o percentual dos votos da pesquisa do data folha
# (Plotting the line chart with the percentage of poll votes from the data sheet)

pesquisas.plot(x='data', y='percentual', color='nome_candidato', markers=True)
```
